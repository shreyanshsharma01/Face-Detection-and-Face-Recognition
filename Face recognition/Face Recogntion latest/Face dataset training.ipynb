```python
import cv2
import os
import numpy as np
from PIL import Image #Python Imaging Library for image editing features 
```

```python
face_recognizer = cv2.face.LBPHFaceRecognizer_create() #using Local Binary Pattern Historgram(LBPH) alogrithm 
face_detector = cv2.CascadeClassifier(r'C:\Users\Q209\Current projects\Face Recognition\haarcascade_frontalface_default.xml')
```

```python
# Path to the face dataset directory
dataset_directory = "face_dataset"
```

```python
# List of individual directories within the face dataset directory
individual_directories = os.listdir(dataset_directory)
```

```python
# Iterate over each individual directory
for individual_directory in individual_directories:
    # Path to the current individual directory
    current_directory = os.path.join(dataset_directory, individual_directory)
    # List of image files within the current individual directory
    image_files = os.listdir(current_directory) #eg: 'shreyansh_14_1.jpg', shreyansh_14_1.jpg'
    faceSamples=[]
    ids = []
    # Iterate over each image file within the current individual directory
    for image_file in image_files:
        # Path to the current image file
        image_path = os.path.join(current_directory, image_file)
        
        PIL_img = Image.open(image_path).convert('L') # convert it to grayscale
        img_numpy = np.array(PIL_img,'uint8')
        id = int(os.path.basename(image_path).split("_")[1]) # to extract id part from 'shreyansh_14_12.jpg' so we split based on '_' and chosse index 1 for middle part 
        faces = face_detector.detectMultiScale(img_numpy)
        
        for (x,y,w,h) in faces:
            faceSamples.append(img_numpy[y:y+h,x:x+w])
            ids.append(id)
    
print ("\n [INFO] Training faces. It will take a few seconds. Wait ...")
```

```python
face_recognizer.train(faceSamples, np.array(ids))   
```

```python
# Save the trained recognizer to an .xml file
face_recognizer.save("Trainer_model/faces_trained.xml")
```
