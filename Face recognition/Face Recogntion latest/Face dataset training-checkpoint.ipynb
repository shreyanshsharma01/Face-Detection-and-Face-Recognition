## Face Training 

```python
def get_images_and_labels(directory_path):
    faceSamples = []
    ids = []
    # Create a dictionary to map user IDs to user names
    id_to_name = {}

    for dir_name in os.listdir(directory_path): 
        subdirectory_path = os.path.join(directory_path, dir_name) #path of individual folder name 
        image_file = os.listdir(subdirectory_path)
        #image_path = os.path.join(subdirectory_path,image_file)
        if not os.path.isdir(subdirectory_path):
            continue

        #label = int(dir_name) 
        for file_name in os.listdir(subdirectory_path):
            image_path = os.path.join(subdirectory_path, file_name)
            individual_id = int(os.path.basename(image_path).split("_")[1])
            #get the user name from directory name
            individual_name = dir_name.split("_")[0]
            id_to_name[individual_id] = individual_name
            
            image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            faceSamples.append(image)
            ids.append(individual_id)
        individual_name = id_to_name[individual_id]
    #return np.array(images), np.array(labels)
    return faceSamples,ids
```

```python
directory_path = "face_dataset" #main directory folder of face dataset
faceSample,ids = get_images_and_labels(directory_path)
face_recognizer.train(faceSample, np.array(ids))
```

```python
# Save the trained recognizer to an .xml file
face_recognizer.save("Trainer_Model/faces_trained.xml")
```
